## Dependencias

```python
import pandas as pd 
import cufflinks as cf
from sklearn.datasets import make_blobs

from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE

cf.go_offline()
```

## Datos sint√©ticos (espacio original $\mathcal{X}$)

```python
X,y = make_blobs(100,6,centers=6,cluster_std=0.1)
```

```python
X= pd.DataFrame(X,columns=[f'x{i+1}' for i in range(6)])
```

```python
X.head()
```

## Cambio de espacio con t-SNE $\mathcal{X}_t$

```python
tsne = TSNE(n_components=3, n_jobs=-1,perplexity=20)

Xt = pd.DataFrame(tsne.fit_transform(X), columns=[f'x{i+1}' for i in range(3)])
```

```python
Xt.iplot(kind='scatter3d',x='x1',y='x2',z='x3',mode='markers',size=10,color='purple')
```
