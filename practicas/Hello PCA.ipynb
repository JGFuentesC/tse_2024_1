## Dependencias

```python
import pandas as pd 
import cufflinks as cf
from sklearn.datasets import make_blobs

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

cf.go_offline()
```

## Datos sintéticos (espacio original $\mathcal{X}$)

```python
X,y = make_blobs(1000,6,centers=6,cluster_std=0.1)
```

```python
X= pd.DataFrame(X,columns=[f'x{i+1}' for i in range(6)])
```

```python
X.head()
```

## Cambio de espacio a componentes principales $\mathcal{Z}$

```python
sc  = StandardScaler()
sc.fit(X)
Xs = pd.DataFrame(sc.transform(X),columns=X.columns)
```

```python
Xs.describe().round(2)
```

```python
pca = PCA(n_components= 3)
```

```python
pca.fit(Xs)
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
Z = pd.DataFrame(pca.transform(Xs),columns=[f'z{i+1}' for i in range(3)])
Z
```

```python
Z.corr().round()
```

```python
Z.iplot(kind='scatter3d',x='z1',y='z2',z='z3',mode='markers',size=10,color='purple')
```
