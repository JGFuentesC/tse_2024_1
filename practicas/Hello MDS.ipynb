## Dependencias

```python
import pandas as pd 
import cufflinks as cf
from sklearn.datasets import make_blobs

from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import MDS

cf.go_offline()
```

## Datos sintéticos (espacio original $\mathcal{X}$)

```python
X,y = make_blobs(100,6,centers=6,cluster_std=0.1)
```

```python
X= pd.DataFrame(X,columns=[f'x{i+1}' for i in range(6)])
```

```python
X.head()
```

## Cambio de espacio con MDS $\mathcal{X}_m$

```python
sc  = MinMaxScaler()
mds = MDS(n_components=3,n_jobs=-1)

Xm = pd.DataFrame(mds.fit_transform(sc.fit_transform(X)),columns=[f'x{i+1}' for i in range(3)])
```

```python
Xm
```

```python
Xm.iplot(kind='scatter3d',x='x1',y='x2',z='x3',mode='markers',size=10,color='purple')
```
